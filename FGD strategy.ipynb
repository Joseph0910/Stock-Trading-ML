```python
coarse_filter,coarse_filter_2,fine_filter = [],[],[]
symbols = pandas.read_csv('tickers_small.csv')
symbols = symbols['Symbol']

for sym in symbols:
  opcf,cash,q_ratio,de_ratio = get_fundamentals(sym)

  # Coarse Filter
  if (opcf <= 0) & (cash < 20000000) & (q_ratio < 1) & (de_ratio >= 5):
    sym2 = sym.replace(" ", "")
    coarse_filter.append(sym2)
  elif (opcf == 1) & (cash == 21000000) & (q_ratio == -1) & (de_ratio == -1):
    sym2 = sym.replace(" ", "")
    coarse_filter_2.append(sym2)
  else:
    pass

for sym in coarse_filter:
  recent_pump, o_n_gaps, above_10, accum, near_52wl = get_technicals(sym)

  # Fine Filter
  if (recent_pump == 0) & (o_n_gaps > 0) & (above_10 == 1) & (accum == 1) & (near_52wl == 1):
    fine_filter.append(sym)

fine_filter
```
